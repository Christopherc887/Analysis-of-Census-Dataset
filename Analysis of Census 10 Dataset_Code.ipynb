Due to the large unique values (1037) in Occupation features, the entries in the Occupation column shall be regrouped into a new Occupation column (category), to facilitate better clarification and visualisation plots. 

```python
# regrouping the entries in the occupation column into a new category
# define a function to categorize the occupation values 

def categorize_occupation(occupation):
    if occupation == 'Unemployed':
        return 'Unemployed'
    elif occupation == 'Child':
        return 'Child'
    elif occupation == 'University Student':
        return 'University Student'
    elif occupation == 'Student':
        return 'Student'
    elif occupation.startswith('Retired'):
        return 'Retired'
    else:
        return 'Employed'

# Create a new column for occupation category
df['Occupation_Category'] = df['Occupation'].apply(categorize_occupation)
```

### Birth Rate (Fertility Rate)

```python
# calculate the birth rate of the population in the census data

# obtain counts of births i.e. ages 0
Number_of_Births = (df['Age'] == 0).sum()

# obtain counts of all women between ages 25-29 
Count_of_Women = ((df["Gender"] == "Female") & (df["Age"] >= 25) & (df["Age"] <= 29)).sum()

# calculate the birth rate 
Birth_Rate = ((Number_of_Births/Count_of_Women) * 1000).round(2)
Birth_Rate 
```

## Mortality 

### Calculation of Crude Death Rate 

The crude death rate can be caluculated by assuming that people aged over 56 years are those likely to pass away in the population, whilst considering the number of deaths in 5 years age bands.

```python
# Calculate the first death rate per annum (ages 56-60)

# obtain the number of people in the age range 56-60 
Number_of_Deaths = (((df["Age"] >= 56) & (df["Age"] <= 60)).sum())

# obtain the total population
Total_Population = len(df)


# Calculate the death rate per annum in the age range 56-60
Death_Rate_per_annum = (((Number_of_Deaths/Total_Population)*1000)/5).round(2)
Death_Rate_per_annum
```

## Determination of the Commuting Level

The commuters are those likely to commute to nearby cities, which include those with Occupation; University Student, PhD Student, Education administrator, Education officer community, Education officer environmental, Education officer museum, Futher education lecturer, Lecturer further education, Higher Education Lecturer, Lecturer higher education, Journalist broadcasting, Journalist magazine, Journalist newspaper, Academic librarian, Librarian academic, Librarian Public, Scientist audiological, Scientist biomedical, Scientist clinical (histocompatibility and immunogenetics), Scientist forensic, Scientist marine, Scientist physiological, Scientist product/process development, Scientist research (life sciences), Scientist research (maths), Scientist research (medical), Scientist research (physical sciences), and Scientist water quality, and also those belonging to Religion; Christian, Muslim, Methodist, Jewish, Sikh, Agnostic, Orthodoxy, Hindu, Bahai and Baptist. 

```python
# calculate the population of individuals by their occupations who are most likely to commute to nearby cities  

# define fuction to group the occupation feature into commuters and non-commuters 

def commuters_occupation(occupation):
    if occupation == 'University Student':
        return 'University Student'
    elif occupation.startswith('Education'):
        return 'Education'
    elif occupation == 'PhD Student':
        return 'PhD Student'
    elif occupation.startswith('Further'):
        return 'Further'
    elif occupation.startswith('Higher'):
        return 'Higher'
    elif occupation.startswith('Journalist'):
        return 'Journalist'
    elif occupation.startswith('Academic'):
        return 'Academic'
    elif occupation.startswith('Lecturer'):
        return 'Lecturer'
    elif occupation.startswith('Librarian'):
        return 'Librarian'
    elif occupation.startswith('Scientist'):
        return 'Scientist'
    else:
        return 'Non Commuter'
    
Commuters_Category = df['Occupation'].apply(commuters_occupation)     
```

```python
# diplays unique values in the new commute category (occupation)
Commuters_Category.unique()
```

```python
# obtain/calculate commuters entries in commute category (occupation) i.e. individuals by occupation in the town who are likely to commute to nearby cities  
Commuters = df.loc[Commuters_Category != 'Non Commuter'].value_counts().sum()  
Commuters
```

```python
# calculate the population of individuals by their religious affiliations who are most likely to commute to nearby cities  

# define fuction to group the religion feature into commuters and non-commuters 

def commuters_religion(religion):
    if religion == 'Christian':
        return 'Commuter'
    elif religion == 'Muslim':
        return 'Commuter'
    elif religion == 'Methodist':
        return 'Commuter'
    elif religion == 'Jewish':
        return 'Commuter'
    elif religion == 'Sikh':
        return 'Commuter'
    elif religion == 'Agnostic':
        return 'Commuter'
    elif religion == 'Orthodoxy':
        return 'Commuter'
    elif religion == 'Hindu':
        return 'Commuter'
    elif religion == 'Bahai':
        return 'Commuter'
    elif religion == 'Baptist':
        return 'Commuter'
    else:
        return 'Non Commuter'
    
Commuters_Category1 = df['Religion'].apply(commuters_religion)     
```

## Determination of Migration Rate 

### Immigration Level

```python
# Calculate the immigration level of the population

# list of people most likely to immigrate are;
# a. Relationship to Head of House 'Visitor'
# b. Relationship to Head of House 'Lodger'
# c. Relationship to Head of House 'None'
# d. Relationship to Head of House 'Head' and Occupation 'University Student'

# Number of immigrants in the population  
Number_of_Immigrants = (df[df["Relationship_to_Head_of_House"] == "Visitor"].value_counts().sum()) + (df[df["Relationship_to_Head_of_House"] == "Lodger"].value_counts().sum()) + (df[df["Relationship_to_Head_of_House"] == "None"].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Head") & (df["Occupation"] == "University Student")].value_counts().sum())
Number_of_Immigrants
```

```python
# Calculate the emmigration level of the population 

# list of people most likely to emmigrate are;
# a. Marital Status 'Divorced'
# b. Relationship to Head of House 'Head' and Occupation 'University Student'
# c. Relationship to Head of House 'None' and Occupation 'University Student'
# d. Relationship to Head of House 'None' and Occupation 'PhD Student'
# e. Relationship to Head of House 'Lodger' and Marital Status 'Divorced'
# f. Relationship to Head of House 'Lodger' and Occupation 'University Student'
# g. Relationship to Head of House 'Lodger' and Occupation 'Student'



# Number of emmigrants in the population
Number_of_Emmigrants = (df[df["Marital_Status"] == "Divorced"].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Head") & (df["Occupation"] == "University Student")].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "None") & (df["Occupation"] == "University Student")].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "None") & (df["Occupation"] == "PhD Student")].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Marital_Status"] == "Divorced")].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Occupation"] == "University Student")].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Occupation"] == "Student")].value_counts().sum())
Number_of_Emmigrants
```

```python
# Calculation Migration Rate

Migration_Rate = (((Number_of_Immigrants - Number_of_Emmigrants)/Population)*1000).round(2)
Migration_Rate
```

A migration rate of -15.6 in a population indicates that more people are leaving the population's geographic area than entering it. In other words, the population is experiencing net out-migration, where the number of emigrants leaving the town exceeds the number of immigrants arriving in the town. This can lead to a decline in the population size and changes in the demographic makeup of the population.
